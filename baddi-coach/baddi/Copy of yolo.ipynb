```python
import yt_dlp

def download_video(url, save_path="/content/"):
    try:
        ydl_opts = {
            'format': 'best',
            'outtmpl': f"{save_path}/%(title)s.%(ext)s",
            'cookiesfrombrowser': ('chrome',)  # Use cookies directly from Chrome browser
        }

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])

        print("Download complete!")
    except Exception as e:
        print("Error:", e)

# Example usage
video_url = "https://www.youtube.com/watch?v=Jb90WzbKRbc"  # Replace with your YouTube URL
download_video(video_url)
```

```python
!pip install opencv-python-headless

import cv2
import matplotlib.pyplot as plt

# Initialize video capture
cap = cv2.VideoCapture("/content/video.mp4")

# Check if video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
else:
    # Read the first frame
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read the first frame.")
        cap.release()
    else:
        # Display the first frame using matplotlib
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.title("First Frame - Select ROI Manually")
        plt.show()

        # Manually input ROI coordinates
        print("Enter ROI coordinates:")
        x = int(input("x: "))
        y = int(input("y: "))
        w = int(input("width: "))
        h = int(input("height: "))
        bbox = (x, y, w, h)

        # Initialize CSRT Tracker
        tracker = cv2.legacy.TrackerCSRT_create()
        tracker.init(frame, bbox)

        # Loop through video frames
        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Update the tracker
            success, bbox = tracker.update(frame)

            # Draw bounding box around the detected object
            if success:
                (x, y, w, h) = [int(i) for i in bbox]
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

            # Display the frame using matplotlib
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.show()

        # Release video capture
        cap.release()
```
